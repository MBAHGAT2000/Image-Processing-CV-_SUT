## Log Transformation

```python
img_float = img.astype(np.float32)

log_img = cv2.log(img_float + 1)

log_img = cv2.normalize(log_img, None, 0, 255, cv2.NORM_MINMAX)

log_img = log_img.astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(log_img, cv2.COLOR_BGR2RGB))
plt.title("Log Transformation")
plt.axis("off")
plt.show()
```

### Log transformation is used to enhance details in dark regions while compressing bright regions, which helps in improving visibility in low-intensity areas.

### Inverse Log Transformation

```python
img_float = img.astype(np.float32) / 255

inv_log_img = cv2.exp(img_float)

inv_log_img = cv2.normalize(inv_log_img, None, 0, 255, cv2.NORM_MINMAX)

inv_log_img = inv_log_img.astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(inv_log_img, cv2.COLOR_BGR2RGB))
plt.title("Inverse Log Transformation")
plt.axis("off")
plt.show()
```

### Enhances the bright regions in the image.

### Power Transformation (Gamma Correction)

```python
#Gamma < 1   →  Brightening
#Gamma = 1   →  No Change
#Gamma > 1   →  Darkening
```

```python
gamma = 2

table = np.array([(i / 255) ** gamma * 255 for i in range(256)]).astype("uint8")

power_img = cv2.LUT(img, table)
```

```python
plt.imshow(cv2.cvtColor(power_img, cv2.COLOR_BGR2RGB))
plt.title("Power Transformation (Gamma Correction)")
plt.axis("off")
plt.show()
```
